```python
train=train.drop('Unnamed: 0', axis=1)
genre_v2=genre_v2.drop('Unnamed: 0', axis=1)
```

```python
datasets = {
    "train": sorted(train.columns),
    "tf": sorted(tf.columns),
    "genre_v2": sorted(genre_v2.columns),
    "genre_excel": sorted(genre_excel.columns)
}

columns_df = pd.DataFrame.from_dict(datasets, orient='index').transpose().fillna('')
columns_df
```

```python
column_rename= {
    'album_name': 'album',
    'artists': 'artist',
    'track_genre': 'genre',
    'track_name': 'name',
    'valence': 'valence',
    'name': 'name', 
    'song_name': 'name', 
    'id': 'track_id',  
    'type': 'track_type',
    'uri': 'track_uri',
    'title':'name'
}

train.rename(columns=column_rename, inplace=True)
tf.rename(columns=column_rename, inplace=True)
genre_v2.rename(columns=column_rename, inplace=True)
genre_excel.rename(columns=column_rename, inplace=True)

```

```python
common_columns = set(train.columns).intersection(tf.columns, genre_v2.columns)

print("Common columns across all datasets:")
common_columns
```

```python
dfs = [train,tf,genre_v2]
unique_columns = {}

for i, df in enumerate(dfs):
    other_columns = set().union(*(dfs[j].columns for j in range(len(dfs)) if j != i))
    unique_columns[f"DataFrame_{i+1}"] = set(df.columns).difference(other_columns)

for df_name, columns in unique_columns.items():
    print(f"Columns unique to {df_name}: {list(columns)}")
```
